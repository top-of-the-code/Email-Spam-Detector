```python
# Basic
import os

# Data handling
import numpy as np

# ML tools
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
import os

emails = []
labels = []

base_path = "data"

folder_mapping = {
    "easy_ham": "ham",
    "hard_ham": "ham",
    "spam_2": "spam"
}

for folder_name, label in folder_mapping.items():
    folder_path = os.path.join(base_path, folder_name)

    print(f"\nChecking folder: {folder_path}")

    for root, dirs, files in os.walk(folder_path):
        print(f"Inside: {root} | Files: {len(files)}")  # DEBUG

        for file in files:
            if file.startswith("."):
                continue

            file_path = os.path.join(root, file)

            try:
                with open(file_path, 'r', encoding='latin-1') as f:
                    content = f.read()
                    
                    if content.strip():   # avoid empty files
                        emails.append(content)
                        labels.append(label)
            except:
                continue

print("\nTotal emails loaded:", len(emails))
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)             # remove numbers
    text = re.sub(r'[^\w\s]', '', text)         # remove punctuation
    return text

emails = [clean_text(email) for email in emails]
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)

X = vectorizer.fit_transform(emails)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y = encoder.fit_transform(labels)

print("Classes:", encoder.classes_)
```

```python
from sklearn.model_selection import train_test_split

X_train_text, X_test_text, y_train, y_test = train_test_split(
    emails, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print(len(X_train_text), len(X_test_text))
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)

X_train = vectorizer.fit_transform(X_train_text)
X_test = vectorizer.transform(X_test_text)
```

```python
model = MultinomialNB(alpha=1.0)
model.fit(X_train_text, y_train)
```

```python
y_pred = model.predict(X_test_text)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nReport:\n", classification_report(y_test, y_pred))
```
